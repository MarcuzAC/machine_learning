```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
salary_data = pd.read_csv(r'C:\Users\User\Desktop\marcus_chithengahhh\machine\ML Algorithms dataset\Salary_Data.csv')
```

```python
x=salary_data.iloc[:, :-1].values
y=salary_data.iloc[:, 1].values
```

```python
sns.barplot(x='YearsExperience', y='Salary', data=salary_data)
sns.set(rc={'figure.figsize':(15,20)})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=1/3, random_state=0)
```

Fitting a linear regression to the training dataset

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train, y_train)
```

```python
ypred = lr.predict(x_test)
ypred
```

```python
plt.scatter(x_test, y_test, color = 'black')
plt.plot(x_train, lr.predict(x_train), color = 'green')
plt.title('Salary VS Experience(train_test)')
plt.ylabel('Years Of Experience')
plt.xlabel('Salary')
plt.show()
```

```python
plt.scatter(x_train, y_train, color = 'black')
plt.plot(x_test, lr.predict(x_test), color = 'green')
plt.title('Salary VS Experience(train_test)')
plt.ylabel('Years Of Experience')
plt.xlabel('Salary')
plt.show()
```
